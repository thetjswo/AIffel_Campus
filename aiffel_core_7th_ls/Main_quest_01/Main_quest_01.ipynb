### 7) 이자지급

- 입금횟수가 5회 될때 : 잔고 기준으로 1% 이자를 잔고에 추가
  - 클래스 생성자에 입금횟수 카운트 변수를 추가하고
  - 입금메소드 호출되면 자동으로 입금횟수 카운트 변수에 1씩 더해주고
  - 입금횟수가 5회가 되면 이자를 더해줌
    - 5의 배수가 될때마다로 조건을 만들어야 5회 이후의 횟수에서도 이자를 더해줄수 있지않을까? => 확인!
