# 키오스크 만들기

- 클래스를 이용해서 키오스크를 생성하고, 가게에서 일어나는 일련의 과정을 메소드로 구현한다.

주문, 추가 주문, 지불, 주문표 작성 메소드를 구현한다

## 1. constructor 생성
- 객체가 만들어지면 메뉴, 가격이 생성되고, 자동으로 주문확인메소드와 메뉴판 메소드를 실행

```python
# Kiosk Class

class Kiosk:
    def __init__(self):
        self.menu = ['ESPRESSO SHOT', 'PECAN CARAMEL', 'SEA SALT CARAMEL', 'ALMOND BUTTER CRUNCH', 'DIPPED APRICOT', 'DEEP DARK']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        
        # say hello
        print("🤍Hello🤍\n🤍Welcome to 🍫Chocolates Makes All Better🍫🤍")
        print("🤍Enjoy the rich, decadent flavors of our VEGAN menu🤍\n")
        
        # check order
        self.check_order()
    
    
    # check order or not
    def check_order(self):
        start = int(input("Wanna ORDER🤍?-- 1\nWanna say GOODBYE😢?-- 2\n"))
        
        # if ready to order
        if start == 1:
            self.menu_print()
            self.menu_select()
        
        # if ready to go home
        elif start == 2:
            print()
            print("🤍Thanks for coming! SEEYA🤍\n🍫Chocolates Makes All Better🍫")
        
        # if value is wrong 
        else:
            print()
            print("🤍TEll ME WHAT YOU WANT🤍")
            self.check_order()  #recursive function 
            
        
    # show menu
    def menu_print(self):
        print()
        print("🍫Chocolates Makes All Better🍫")
        print()
        print("🤍Here's OUR MENU🤍")
        
        # make index with enumerate function
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', '₩', self.price[i])
```

## 2. 주문받기 & 메뉴선택
- menu_select 메소드 실행후 자동으로 지불, 주문서 출력 메소드 실행

```python
    # get orders
    def menu_select(self):
        print() # 한 줄 띄우기
        self.order_menu = [] # order list
        self.order_price = [] # price list

        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("🤍What's your choie?🤍\n🤍Choose number🤍: ")) # get order

            # if n is in menu number
            if 1 <= n & n <= len(self.menu):
                self.order_price.append(self.price[n-1])
                self.price_sum = self.price[n-1]

            # if n is out of menu number
            else:
                print("The menu doesn't exist. Please reorder.")

        # check here or togo
        t = 0
        # ask t is right value
        while t != 1 and t != 2:
            t = int(input("for HOT => 1, for ICE => 2 : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("Please enter either 1 or 2.\n")

        # add in order list and price list
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])
        print(self.order_menu[0], " : ", "₩", self.price_sum)

        # check extra order or ready to pay
        while n != 0:
            print()
            n = int(input("For Additional Orders => Enter chocolate number\nFor payment => 0 : "))
            # get extra order
            if n > 0 and n < len(self.menu)+1:
                #check temperature
                t = 0
                while t != 1 and t != 2:
                    t = int(input("for HOT => 1, for ICE => 2 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("Please enter either 1 or 2.\n")
                # add extra ordered menu in order menu list
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                # add extra ordered menu price in price list
                self.order_price.append(self.price[n-1])
                # set extra ordered menu price in 'price_sum' variable
                self.price_sum = self.price[n-1]
                # print extra ordered menu and total price
                print("Additional orders\n", self.temp, self.menu[n-1], ":", "₩", self.price_sum, '\n', 'Total : ', "₩", sum(self.order_price))
            else:
                if n == 0: # ready to pay
                    print("Your order is completed.")
                    print(self.order_menu, self.order_price) # show total orderlist
                else: # get the others' value
                    print("The menu doesn't exist. Please reorder.")
                    
        self.table()
        self.pay()
```

## 3. 지불, 주문서 출력

### 지불
- 현금과 카드 선택지 제시
- input은 값을 str으로 받기 때문에 숫자 1, 2를 입력하더라고 조건문에서는 str으로 받아서 코드를 단순화
- 올바른 값이 들어오지 않을 경우 재귀함수로 다시 실행

### 주문서 출력
- 전체 메뉴를 다시 한번 출력

```python

    # 지불
    def pay(self):
        print()
        print(f"Total cost is {self.order_price}")
        n = input("현금('cash' or 1)\n카드('card' or 2)\n")
        if n == 'cash' or n == '1':
            print('직원을 호출하겠습니다.')
        elif n == 'card' or n == '2':
            print('IC칩 방향에 맞게 카드를 꽂아주세요.')
        else:
            print("다시 결제를 시도해 주세요.")
            self.pay()

    # 주문서 출력
    def table(self):
        # 외곽 line
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print('\t', self.order_menu[i], ' : ', self.order_price[i])

        print('\t'*2, 'Total :', sum(self.order_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

```
